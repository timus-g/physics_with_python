## Some Plottings

Line Plots

```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
```

```python
plt.plot(x,y)
```

Plot colour and line marker size

```python
plt.plot(x,y,'o--',color="purple", lw=0.7, ms=5)
```

Figure size

```python
plt.figure(figsize=(8,3))
plt.plot(x,y,'o--',color="purple", lw=0.7, ms=5)
```

Axis labels

```python
plt.figure(figsize=(8,3))
plt.plot(x,y,'o--',color="purple", lw=0.7, ms=5)

plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
```
